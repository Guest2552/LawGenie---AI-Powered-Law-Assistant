```python
import openai
import gradio as gr
import fitz  # PyMuPDF
import docx

# Set your OpenAI API key
openai.api_key = 'sk-proj-cJ7b3-JbuGrf9gD2WmrYI6n_oeb4GYhzO788a6uW_KYbvWVrSE78edtxTEdv7s3yN93IP8A1-T3BlbkFJF8DDHyoGqVWUT8BAWyBw48JvO4dMd-ZtseM-p0DodFDnODy68oIWoaQezrXL__7A9cA-ZFVNAA'

# ---- Document Readers ----
def read_pdf(file_path):
    text = ""
    with fitz.open(file_path) as pdf:
        for page in pdf:
            text += page.get_text()
    return text.strip()

def read_docx(file_path):
    doc = docx.Document(file_path)
    return "\n".join([para.text for para in doc.paragraphs]).strip()

def extract_text(file):
    if file.name.endswith(".pdf"):
        return read_pdf(file.name)
    elif file.name.endswith(".docx"):
        return read_docx(file.name)
    else:
        return "❌ Unsupported file format. Please upload PDF or DOCX."

# ---- Main Chat Function ----
def chat_with_case(history, user_input, uploaded_file):
    try:
        # Extract file text if uploaded
        case_text = ""
        if uploaded_file is not None:
            case_text = extract_text(uploaded_file)
            if case_text.startswith("❌"):
                history.append((user_input, case_text))
                return history, ""

        # Combine prompt
        combined_prompt = f"User question: {user_input}\n"
        if case_text:
            combined_prompt += f"\nCase Document Content:\n{case_text}\n\nBased on the above case file and question, explain the legal situation clearly."

        # Typing animation
        history.append((user_input, "🕐 Typing..."))

        messages = [
            {"role": "system", "content": "You are a helpful legal assistant skilled in Indian law. Answer based on the case file if provided."},
            {"role": "user", "content": combined_prompt}
        ]

        response = openai.ChatCompletion.create(
            model="gpt-4",
            messages=messages,
            temperature=0.6,
            max_tokens=1200
        )

        reply = response['choices'][0]['message']['content'].strip()
        history[-1] = (user_input, reply)
        return history, ""
    except Exception as e:
        history[-1] = (user_input, f"⚠️ Error: {str(e)}")
        return history, ""

# ---- UI Starts Here ----
emblem_path = "Emblem_of_India.png"

with gr.Blocks(title="LawGenie: Indian Legal Chatbot") as demo:
    # Header with emblem and gradient banner
    with gr.Row():
        with gr.Column(scale=0):
            gr.Image(value=emblem_path, show_label=False, width=100)
        with gr.Column(scale=5):
            gr.HTML("""
            <div style="
                background: linear-gradient(to right, #FF9933, #FFFFFF, #138808);
                padding: 20px;
                border-radius: 10px;
                text-align: center;
                box-shadow: 0 0 10px rgba(0,0,0,0.2);
                margin-bottom: 10px;
            ">
                <h1 style="font-size: 2.5em; margin: 0; font-family: 'Segoe UI', sans-serif;">
                    LawGenie: AI-Powered Legal Assistant
                </h1>
                <p style="font-size: 1.1em; color: #333;">
                    Get instant Indian legal advice powered by GPT-4
                    Upload your case file (PDF/DOCX) and ask any legal question.
                </p>
            </div>
            """)

    # Chatbot window
    chatbot = gr.Chatbot(label="Chat with LawGenie", height=450, show_copy_button=True)

    # Input Row: Textbox + File Upload
    with gr.Row():
        with gr.Column(scale=5):
            msg = gr.Textbox(
                placeholder="Type your legal question and press Enter...",
                label="Your Question",
                lines=1
            )
        with gr.Column(scale=2):
            upload = gr.File(
                label="Upload Case File",
                file_types=[".pdf", ".docx"]
            )

    # Clear button below textbox
    with gr.Row():
        with gr.Column(scale=5):
            clear = gr.Button("🧹 Clear Chat")

    # Function bindings
    msg.submit(fn=chat_with_case, inputs=[chatbot, msg, upload], outputs=[chatbot, msg])
    clear.click(fn=lambda: [], inputs=None, outputs=chatbot)

    # ---- Auto-scroll Script ----
    gr.HTML("""
    <script>
        const observer = new MutationObserver(() => {
            const containers = document.querySelectorAll('[data-testid="chatbot"]');
            containers.forEach(container => {
                container.scrollTop = container.scrollHeight;
            });
        });
        observer.observe(document.body, { childList: true, subtree: true });
    </script>
    """)

# ---- Run App ----
if __name__ == "__main__":
    demo.launch()
```
